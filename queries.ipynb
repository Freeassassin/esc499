```python
# Load data from .dat files into TPC-DS tables
# The data files use '|' as delimiter with a trailing delimiter

import os

# Mapping of table names to their data files
TABLES = [
    "call_center",
    "catalog_page", 
    "catalog_returns",
    "catalog_sales",
    "customer",
    "customer_address",
    "customer_demographics",
    "date_dim",
    "dbgen_version",
    "household_demographics",
    "income_band",
    "inventory",
    "item",
    "promotion",
    "reason",
    "ship_mode",
    "store",
    "store_returns",
    "store_sales",
    "time_dim",
    "warehouse",
    "web_page",
    "web_returns",
    "web_sales",
    "web_site",
]

# Load each table
for table in TABLES:
    file_path = os.path.join(DATA_DIR, f"{table}.dat")
    if os.path.exists(file_path):
        try:
            # Check if table already has data
            existing_count = con.execute(f"SELECT COUNT(*) FROM {table}").fetchone()[0]
            if existing_count > 0:
                print(f"Skipping {table} - already has {existing_count:,} rows")
                continue
            
            # Get column count from the table
            col_count = con.execute(f"""
                SELECT COUNT(*) 
                FROM information_schema.columns 
                WHERE table_name = '{table}'
            """).fetchone()[0]
            
            # Generate column names for read_csv (col0, col1, ... + dummy for trailing |)
            col_names = [f"column{i}" for i in range(col_count + 1)]  # +1 for trailing delimiter
            
            # Read CSV with auto-detection, selecting only needed columns
            con.execute(f"""
                INSERT INTO {table} 
                SELECT * EXCLUDE (column{col_count})
                FROM read_csv('{file_path}', 
                    delim='|', 
                    header=false,
                    names={col_names},
                    null_padding=true
                )
            """)
            row_count = con.execute(f"SELECT COUNT(*) FROM {table}").fetchone()[0]
            print(f"Loaded {row_count:,} rows into {table}")
        except Exception as e:
            print(f"Error loading {table}: {e}")
    else:
        print(f"Warning: {file_path} not found")

print("\nData loading complete!")
```

```python
# Verify the data loading by showing table counts
print("TPC-DS Table Summary:")
print("-" * 40)
for table in TABLES:
    try:
        count = con.execute(f"SELECT COUNT(*) FROM {table}").fetchone()[0]
        print(f"{table:30} {count:>10,}")
    except Exception as e:
        print(f"{table:30} {'Error':>10}")
print("-" * 40)
```

## Run queries

```python
# Load and run all TPC-DS queries
import time

QUERIES_FILE = "./DSGen-software-code-4.0.0/tools/queries/query_0.sql"

# Read and parse queries
with open(QUERIES_FILE, 'r') as f:
    content = f.read()

# Split by semicolon and filter empty queries
queries = [q.strip() for q in content.split(';') if q.strip()]

print(f"Found {len(queries)} queries to execute\n")

# Execute each query and track timing
results = []
for i, query in enumerate(queries, 1):
    try:
        start_time = time.time()
        result = con.execute(query).fetchall()
        elapsed = time.time() - start_time
        row_count = len(result)
        results.append({
            'query': i,
            'status': 'OK',
            'rows': row_count,
            'time_ms': elapsed * 1000
        })
        print(f"Query {i:3d}: {row_count:>8,} rows in {elapsed*1000:>8.2f} ms")
    except Exception as e:
        elapsed = time.time() - start_time
        results.append({
            'query': i,
            'status': 'ERROR',
            'rows': 0,
            'time_ms': elapsed * 1000,
            'error': str(e)
        })
        print(f"Query {i:3d}: ERROR - {str(e)[:60]}...")

# Summary
successful = sum(1 for r in results if r['status'] == 'OK')
total_time = sum(r['time_ms'] for r in results)
print(f"\n{'='*50}")
print(f"Summary: {successful}/{len(queries)} queries succeeded")
print(f"Total execution time: {total_time/1000:.2f} seconds")
```
